### 42. Check if a number is perfect (sum of factors equals number).

```python
num = int(input("Enter a number: "))

total = 0

for i in range(1, num):
    if num % i == 0:
        total += i

if total == num:
    print("sum of factors equals number")
else:
    print("sum of factors not equals number")
```
