# Basic List Operations 

1. Create a list of 5 numbers and add a new number to the end.


```python
lst=[10,20,30,40,50]

print("old list ->",lst)

# add new number at the end
lst.append(60)

print("New list ->",lst)
```

2. Make a list of fruits and insert a fruit at the second position.


```python
fruits=["Apple","Banana","Kiwi","Orange"]
print(fruits)

# insert a fruit "Mango" at the second index position
fruits.insert(2,"Mango")
print(fruits)
```
